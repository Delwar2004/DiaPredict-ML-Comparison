Import the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Analysis

PIMA Diabetes Dataset

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
#printing the first 5 rows of the datset
diabetes_dataset.head()
```

```python
# number of rows and column in this diabetes dataset
diabetes_dataset.shape
```

```python
#getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
# checking the missing value
diabetes_dataset.isnull().sum()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0-->Non-Diabetec
1--> Diabetec

```python
# finding the mean(all columns) for 0 and 1 
diabetes_dataset.groupby('Outcome').mean()
```

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X,Y)
```

Splitting the data into Training data and Test Data

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y ,test_size = .2, random_state =2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

Model Evaluation

Accuracy score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data: ',training_data_accuracy)
```

```python
#accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data: ',test_data_accuracy)
```

Making a predictive system

```python
# Order: Pregnancies, Glucose, BP, SkinThickness, Insulin, BMI, Pedigree, Age
input_data = (4, 110, 92, 0, 0, 37.6, 0.191, 30) # these data refers a non-diabetec person

# Convert input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#  Reshape the array for a single instance prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#  Standardize the input data using the loaded scaler
std_data = scaler.transform(input_data_reshaped)

#  Make prediction using the loaded model
prediction = classifier.predict(std_data)

#  Display the final result
if (prediction[0] == 0):
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
